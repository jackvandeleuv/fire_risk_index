```python
public['std_addr'] = public.std_addr.str.upper()
public['std_st'] = public.std_st.str.upper()
public['std_city'] = public.std_city.str.upper()
```

```python
def correct_address(std_addr):
    if pd.isna(std_addr):
        return np.nan, np.nan
    if 'APT' in std_addr:
        split_addr = std_addr.split('APT')
        return split_addr[0].strip(), split_addr[1].strip()
    if 'UNIT' in std_addr:
        split_addr = std_addr.split('UNIT')
        return split_addr[0].strip(), split_addr[1].strip()
    if 'SUB' in std_addr:
        split_addr = std_addr.split('SUB')
        return split_addr[0].strip(), split_addr[1].strip()
    if '#' in std_addr:
        split_addr = std_addr.split('#')
        return split_addr[0].strip(), split_addr[1].strip()
    if 'LOT' in std_addr:
        split_addr = std_addr.split('LOT')
        return split_addr[0].strip(), split_addr[1].strip()
    return std_addr, np.nan

public[['corrected_address', 'corrected_unit']] = public.std_addr.apply(correct_address).apply(pd.Series)
```

```python
public['corrected_address'] = public.corrected_address.str.replace('OFC', '')
public['corrected_address'] = public.corrected_address.str.replace('1SUD', '')
```

```python
public.corrected_unit = public.corrected_unit.fillna('')
# df.apt_no = df.apt_no.fillna('')
```

```python
conn = sqlite3.Connection('./data/nfirs/' + 'fire_data.db')
cur = conn.cursor()

cur.execute('DROP TABLE IF EXISTS public_housing_addresses')
cur.execute("""CREATE TABLE IF NOT EXISTS public_housing_addresses (
    ID INTEGER PRIMARY KEY,
    corrected_address TEXT,
    corrected_unit TEXT,
    std_city TEXT,
    std_st TEXT,
    std_zip5 INTEGER
)""")
conn.commit()

addresses = public.loc[:, ['corrected_address', 'corrected_unit', 'std_city', 'std_st', 'std_zip5']]
addresses = addresses.drop_duplicates(keep='first')
addresses.to_sql('public_housing_addresses',
    conn, 
    if_exists='append', 
    index=False
)
conn.commit()

cur.execute("""
    CREATE INDEX idx_pub_housing 
    ON public_housing_addresses(corrected_address, corrected_unit, std_city, std_st, std_zip5)
""")
conn.commit()
conn.close()
```
