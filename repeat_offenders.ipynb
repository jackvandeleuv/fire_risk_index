250 buildings have lost over 90% of the total money lost in fires

```python
prop_cols = [x for x in df.columns if 'prop_loss_' in x]
cont_cols = [x for x in df.columns if 'cont_loss_' in x]
df['total_prop_loss'] = df[prop_cols].sum(axis=1)
df['total_cont_loss'] = df[cont_cols].sum(axis=1)
df['total_money_lost'] = df.total_prop_loss + df.total_cont_loss
df = df.sort_values('total_money_lost', ascending=False)
df['cumulative_money_lost'] = df['total_money_lost'].cumsum()
df['cumulative_percent_money_lost'] = df.cumulative_money_lost / df.total_money_lost.sum()
df.cumulative_percent_money_lost.head(230)
```

```python
df[cont_cols].sum()
```

```python
len(df[df.total_money_lost > 0]) / len(df)
```

150 out of about 150,000 buildings have had about half of the identified fires.

```python
fire_cols = [x for x in df.columns if 'building_fire' in x]
df['total_fires'] = df[fire_cols].sum(axis=1)
df = df.sort_values('total_fires', ascending=False)
df['cumulative_fires'] = df['total_fires'].cumsum()
df['cumulative_percent_fires'] = df.cumulative_fires / df.total_fires.sum()
df.cumulative_percent_fires.head(150)
```

About 1000 buildings caused 80% of fires.
